```python
import pandas as pd
```

```python
file_path = ('codebar-AI-fundamentals-pixels-banking-ML-model/data/pixels_banking_loan_repayment_capstone_dataset.csv')

print('Loading file:', file_path)
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
missing_values = df.isnull().sum()
missing_values[missing_values > 0].sort_values(ascending=False)
```

```python
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_percent[missing_percent > 0].sort_values(ascending=False)
```

```python
clean_df = df.copy()

clean_df = clean_df.drop(columns=['customer_id'], errors='ignore')

clean_df.head()
```

```python
numeric_columns = clean_df.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = clean_df.select_dtypes(include=['object']).columns

print('Numeric columns:')
print(list(numeric_columns))

print('Categorical columns:')
print(list(categorical_columns))
```

```python
flag_columns = ['credit_score', 'monthly_income']
for column in flag_columns:
    clean_df[column + '_was_missing'] = clean_df[column].isnull().astype(int)
```

```python
for column in numeric_columns:
    median_value = clean_df[column].median()
    clean_df[column] = clean_df[column].fillna(median_value)
```

```python
for column in categorical_columns:
    clean_df[column] = clean_df[column].fillna("Unknown")
```

```python
clean_df.isnull().sum().sum()
```

```python
for column in categorical_columns: 
    print(column)
    print(clean_df[column].value_counts())
    print('---')
```

```python
df_encoded = pd.get_dummies(
    clean_df,
    columns=categorical_columns,
    drop_first=True
)

df_encoded.head(50)
```

```python
print('Original shape:', df.shape)
print('Prepared shape before encoding:', clean_df.shape)
print('Shape after encoding:', df_encoded.shape)
```

```python
list(df_encoded.columns)
```

```python
df_encoded.isnull().sum().sum()
```

```python
df_encoded.select_dtypes(include='object').columns.tolist()
```

```python
df_model_ready = df_encoded.copy()

df_model_ready.head()
```

```python
print('Missing values left:', df_model_ready.isnull().sum().sum())
print('Text columns left:', list(df_model_ready.select_dtypes(include='object').columns))
print('Final shape:', df_model_ready.shape)
print('Customer ID still present:', 'customer_id' in df_model_ready.columns)
```

```python
clean_file_path = 'codebar-AI-fundamentals-pixels-banking-ML-model/data/pixels_banking_loan_repayment_clean.csv'

# 2. Save it
print('Saving clean data to:', clean_file_path)
df_model_ready.to_csv(clean_file_path, index=False)
```
